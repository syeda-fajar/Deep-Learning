**Resizing Image Using Pillow**

```python
from PIL import Image
```

```python
img = Image.open('/content/photo-1507146426996-ef05306b995a?fm=jpg&q=60&w=3000&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxzZWFyY2h8Mnx8cHVwcHl8ZW58MHx8MHx8fDA=')
img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# loading an image through matplotlib.image module
img_res=mpimg.imread('/content/dog_image_resized.jpg')
img_resize_plot = plt.imshow(img_res)
plt.show()
```

## **Converting RGB images to grayscale using opencv**

```python
import cv2
```

```python
img = cv2.imread('/content/photo-1507146426996-ef05306b995a?fm=jpg&q=60&w=3000&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxzZWFyY2h8Mnx8cHVwcHl8ZW58MHx8MHx8fDA=')
```

```python
type(img)
```

```python
img.shape
```

```python
gryscale = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(gryscale)
```

```python
gryscale.shape
```
